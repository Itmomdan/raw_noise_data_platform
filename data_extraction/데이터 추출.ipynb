## 데이터 추출

```python
import config 
from config import DB_CONFIG
import pandas as pd
import pymysql  

# MySQL 연결
conn = pymysql.connect(
    host=DB_CONFIG['host'],
    user=DB_CONFIG['user'],
    password=DB_CONFIG['password'],
    database=DB_CONFIG['database'],
    port=DB_CONFIG['port']
)

# 데이터 불러오기
queries = {
    "car_siren": "SELECT * FROM merge50_car_siren",
    "car_horn": "SELECT * FROM merge50_car_horn",
    "car_driving": "SELECT * FROM merge50_car_driving",
    "motorcycle_horn": "SELECT * FROM merge50_motorcycle_horn",
    "motorcycle_driving": "SELECT * FROM merge50_motorcycle_driving"
}

dfs = {name: pd.read_sql(query, conn) for name, query in queries.items()}

# MySQL 연결 종료
conn.close()

# 데이터 병합
df_combined = pd.concat(dfs.values(), ignore_index=True)

# CSV 저장
for name, df in dfs.items():
    df.to_csv(f"{name}.csv", index=False, encoding='utf-8-sig')

df_combined.to_csv("combined.csv", index=False, encoding='utf-8-sig')

print("CSV 파일 저장 완료!")
```
